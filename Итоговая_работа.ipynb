## Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
df_high = df[df.salary == 'high'].reset_index(drop= True)
df_high
```

```python
df_low = df[df.salary == 'low'].reset_index(drop= True)
df_low
```

```python
sns.kdeplot(df_high['average_montly_hours'].values, label="high", fill = True, color = '#000000')
sns.kdeplot(df_low['average_montly_hours'].values, label="low", fill = True, color = '#B22222')
plt.legend()
plt.show()
```

```python
from scipy.stats import shapiro, mannwhitneyu
```

```python
stat, p = shapiro(df_high['average_montly_hours'].values)

print(f'stat = {stat:.3f}, p = {p:.5f}')
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

```python
stat, p = shapiro(df_low['average_montly_hours'].values)

print(f'stat = {stat:.3f}, p = {p:.5f}')
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

```python
stat, p = mannwhitneyu(df_high['average_montly_hours'].values, df_low['average_montly_hours'].values)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

Вывод: гипотеза не подтверждается. Нет различий в проведенном на работе времени от уровня зарплаты.
